```python

katok=[]


#데이터를 중간에 삽입하는 plu()함수
def plu(num,name):
    if num<=0 or num>len(katok):
        print("데이터 범위를 벗어났습니다")
        return
    
    katok.append(None)                         #빈 공간 생성
    
    for i in range (len(katok)-1,num-1,-1):    #마지막부터,원하는 위치까지,-1 증가량
        katok[i]=katok[i-1]                     #
        katok[i-1]=None
        
    katok[num-1]=name                           #for문이 끝나고 지정한 위치에 친구 추가
    
#데이터를 삭제하는 ma()함수
def ma(num):
    
    if num<=0 or num>len(katok):
        print("데이터 범위를 벗어났습니다")
        return 
    
    katok[num-1]=None
    for i in range (num,len(katok),1):
        katok[i-1]=katok[i]
        katok[i]=None
    del(katok[len(katok)-1])    #마지막 칸 제거
    
    
    
    
select=0 #선택변수

while(select!=3):
    
    select=int(input("선택하세요(1.추가 2.삭제 3.종료)-->"))
    
    if (select==1):
        add =input("추가 할 데이터-->").split('",')   #두개 받기
        name = add[0]
        num = add[1]
        
        realname = name.lstrip("(\"")
        realnum = num.rstrip(')')
        realrealnum = int(realnum)
        katok.append((realname, realrealnum))
        
        katok.sort(key = lambda x:x[1], reverse=True)
        print(katok)
  
    elif (select==2):
        pos=int(input("삭제할 위치-->"))
        ma(pos)
        print(katok)
        
    elif (select==3):
        print(katok)
        exit
    else:
        print("1~3 중 하나를 선택하세요")
        continue
```
